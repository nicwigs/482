# Takes list of stock in nasdaq and then creates smaller lists to retrieve data 
-List of stocks from  https://www.nasdaq.com/screening/company-list.aspx!

-Stock data from https://finance.yahoo.com/

-Stock data in bulk from these small lists - http://finance.jasonstrimpel.com/bulk-stock-download/

```python
import pandas as pd
```

```python
d = pd.read_csv('nasdaqList.csv')
```

```python
d
```

```python
stock = d['Name'].tolist()
```

```python
n = 100
b = [stock[i:i + n] for i in range(0, len(stock), n)]
len(b)
```

```python
f = open('List1.csv', "w")
for l in b[0]:
  f.write(l)
  f.write("\n")
f.close()


```

```python
for i in range(len(b)):
    n = 'List'+str(i)+'.csv'
    f = open(n, "w")
    for l in b[i]:
      f.write(l)
      f.write("\n")
    f.close()
```
