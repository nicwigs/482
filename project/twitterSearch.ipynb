## Just Used to look at recent tweets, get a feel for API

```python
import pandas as pd
```

```python
import tweepy
from tweepy import OAuthHandler
from tweepy import API
import config

C_KEY = config.C_KEY
C_SECRET = config.C_SECRET
A_TOKEN_KEY = config.A_TOKEN_KEY
A_TOKEN_SECRET = config.A_TOKEN_SECRET

auth = tweepy.OAuthHandler(C_KEY, C_SECRET)
auth.set_access_token(A_TOKEN_KEY, A_TOKEN_SECRET)
```

```python
api = tweepy.API(auth)

#keyword = 'tesla from:elonmusk'
keyword = 'tesla lang:en'

posts = api.search(q=keyword, count=100)
#posts = api.search(q=keyword)
```

```python
posts
```

```python
#d = pd.DataFrame(columns=['date','text'])
d = []
```

```python
posts[0].created_at.strftime("%d/%m/%Y")
```

```python
for tweet in posts:
    d.append({'date':tweet.created_at.strftime("%d/%m/%Y"), 'text':tweet.text})
    
d
```

```python
len(d)
```

```python
import json
a = json.dumps(d)
```

```python
with open('tweets0324.json', 'w') as outfile:
    json.dump(a, outfile)
```
