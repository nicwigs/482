```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
sa = pd.read_csv("../../project/data/opinion/SentimentAnalysisDataset.csv",error_bad_lines=False)
```

```python
sa
```

```python
sa['bad'] = sa.SentimentText.apply(lambda x: '\n' in x)
```

```python
# sometimes csv reads wrong - and then Sentiment text has a new line which screws up
sa = sa[sa.bad == False]
```

```python
sa[sa.bad]
```

```python
X = sa.SentimentText.tolist()
y = sa.Sentiment.tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
train = pd.DataFrame(X_train,y_train)
train
```

```python
train.to_csv('../../project/data/opinion/train.data', sep='\t',header=False)
```
