```python
import json
import pandas as pd
import os
import numpy as np
```

```python
directory = '../../project/data/twitter/1516/'
paths = []
for root, dirs, files in os.walk(directory):
    for file in files:
        paths.append(directory+file)
```

```python
paths
```

```python
data = []
for path in paths:
    with open(path) as f:  
        data += json.loads(json.load(f))
```

```python
len(data)
```

```python
d2 = pd.DataFrame(columns=['created_at','text'])
```

```python
for tweetJson in data:
    d2 = d2.append(tweetJson, ignore_index=True)
```

```python
d2
```

```python
d3 = d2.copy()
```

```python
d3['date'] = d3.apply(lambda x: pd.to_datetime(x.date).date(), axis = 1)
```

```python
d3
```

```python
g1 = d3.groupby('date')
```

```python
dayStocks = g1.count()
dayStocks
```

```python
g1.count().created_at.agg([np.sum, np.mean, np.std])
```

```python
p = dayStocks.plot(use_index=True, y='created_at')
p
```
