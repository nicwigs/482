# Puts together all the stock data recieved, removes NaN columns


```python
import pandas as pd
import os
```

```python
directory = '../../project/data/stock/results/'
paths = []
for root, dirs, files in os.walk(directory):
    for file in files:
        paths.append(directory+file)
```

```python
data = pd.read_csv(paths[0])
data = data.set_index('Date')
for path in paths[1:]:
    with open(path) as f:  
        temp = pd.read_csv(path)
        temp = temp.set_index('Date')
        data = data.copy().join(temp)
```

```python
data
```

```python
d = data.copy()
d = d.reset_index()
d.Date = pd.to_datetime(d.Date)
d
```

```python
# twitter data start 2015-04-27, 2016-04-06
tweetStartDate = pd.to_datetime('2015-04-27')
tweetEndDate = pd.to_datetime('2016-04-06')

offset = 7

# classifier will look at last few days of stock as well -> hence offset
# Upper is one more than when our tweets stop, since tweets will be offset by 1 day -> use todays tweets to 
# predict tomorrow 
lower = (tweetStartDate - pd.Timedelta(days=offset)).date()
upper = (tweetEndDate + pd.Timedelta(days=1)).date()

d2 = d.copy()[(d.Date >= lower) & (d.Date <= upper) ]
```

```python
d2
```

```python
d3 = d2.dropna(axis=1)
```

```python
d3 = d3.set_index('Date')
```
